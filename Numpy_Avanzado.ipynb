## 19 - ¿Cómo obtener las fechas de ayer, hoy y mañana?

`pista: np.datetime64, np.timedelta64`


```python
today = np.datetime64('today')
yesterday = np.datetime64('today')-1
tomorrow = np.datetime64('today')+1
```

```python
# Obtener la fecha de hoy
hoy = np.datetime64('today', 'D')

# Calcular las fechas de ayer y mañana
ayer = hoy - np.timedelta64(1, 'D')
manana = hoy + np.timedelta64(1, 'D')

# Mostrar las fechas
print("Fecha de ayer:", ayer)
print("Fecha de hoy:", hoy)
print("Fecha de mañana:", manana)
```

```python
from datetime import datetime, timedelta

# Obtener la fecha de hoy
hoy = datetime.now().date()

# Calcular las fechas de ayer y mañana
ayer = hoy - timedelta(days=1)
mañana = hoy + timedelta(days=1)

# Mostrar las fechas
print("Fecha de ayer:", ayer)
print("Fecha de hoy:", hoy)
print("Fecha de mañana:", mañana)
```

```python
A = np.random.randint(0,2,5)
B = np.random.randint(0,2,5)
print(A)
print(B)
# Si son identicas en forma pero no tienen los mismos valores (hay tolerancia).
equal = np.allclose(A,B)
print(equal)

# Si son identicas en forma y además tienen los mismos valores (no hay tolerancia).
equal = np.array_equal(A,B)
print(equal)
```

## 30 - ¿Cómo convertir una matriz `float` (32 bits) en tipo `int` (32 bits)?

`pista: view and [:] =`

```python
# para saber de cuanto tengo que hacer la matriz, tengo que hacer el calculo de los espacios. 
# cada float son 4 bytes. asi que seran 32/4 = 8. 

float_matrix = np.arange(1, 9, dtype=np.float32).reshape(4, 2)
float_matrix

#ahora lo paso a int. para ello se usa view. el simbolo [:] vale para guardarlo como nueva vista sin afectar a float_matrix
int_matrix = float_matrix.view(np.int32)[:]
int_matrix
```

## 31 - ¿Cuál es el equivalente de `enumerate` para matrices en `numpy`?

`pista: np.ndenumerate, np.ndindex`

```python
Z = np.arange(16).reshape(4,4)
for index, value in np.ndenumerate(Z):
    print(index, value)
for index in np.ndindex(Z.shape):
    print(index, Z[index])
```

## 32 - Restar la media de cada fila a una matriz.

`pista: mean(axis=, keepdims=)`


```python

# Crear una matriz de ejemplo
X = np.array([[10, 20, 30],
                   [40, 50, 60],
                   [70, 80, 90]])

# Calcular la media de cada fila
row_means = X.mean(axis=1, keepdims=True)

# Restar la media de cada fila a la matriz
result = X - row_means

print("Matriz original:")
print(X)

print("\nMedia de cada fila:")
print(row_means)

print("\nMatriz después de restar la media de cada fila:")
print(result)
```

## 34 - ¿Cómo saber si una matriz 2D tiene columnas nulas?

`pista: any, ~`


```python
# Crear una matriz 2D de ejemplo
matrix = np.array([[1, 2, 0],
                   [4, 5, 0],
                   [7, 8, 0]])

# Comprobar si hay columnas nulas (todas las filas son cero en la columna)
has_null_columns = np.any(np.all(matrix == 0, axis=0))

# Mostrar el resultado
print("Matriz:")
print(matrix)
print("¿Hay columnas nulas?", has_null_columns)


# Explicación del código:
# matrix == 0: Crea una matriz booleana donde cada elemento es True si es igual a 0 y False en caso contrario.
# np.all(..., axis=0): Verifica si todas las filas de cada columna son 0. Esto devuelve un array booleano que indica qué columnas son nulas (True si la columna es nula, False si no).
# np.any(...): Comprueba si hay al menos un valor True en el array booleano resultante, lo que indica que hay al menos una columna nula.

```
